## 3 Pipline: Crear conjunto de prueba

#### 3.1.2 Variables categóricas

Esto es más artesanal. Me voy a ayudar con boxplots.

##### 3.1.2.1 conclusión

Esto garantiza que ambos conjuntos sean representativos de los dos grandes segmentos de precio identificados en el análisis exploratorio, evitando sesgos.

#### 3.1.3 Conclusión de las correlaciones

- La variable categórica más importante es "Transmission". -> la que me importa para el muestreo estratificado (Stratified Sampling).

## 5 Preparar los datos para los algoritmos de Machine Learning

**Obs.** Cuando defino num_attribs, no le digo a Scikit-Learn "esto ya es un número", le digo "mandá estas columnas por la tubería que procesa los números". 

La clave está en el orden de los pasos dentro del num_pipeline. Para que no rompa nada, el primer eslabón de la cadena tiene que ser el cleaner.

```python
from src.data_split import split_train_test_stratified
from src.pipeline import build_full_pipeline

#Separo etiquetas ("Price") de predictores
cars_train = train_set.drop("Price", axis=1)
cars_labels_train = train_set["Price"].copy()

#Defino qué columnas van a cada pipeline
num_attribs = ["Year", "Kilometer", "Engine", "Max Power", "Max Torque", "Seating Capacity", "Fuel Tank Capacity"]
cat_attribs = ["Fuel Type", "Transmission", "Drivetrain", "Make"]

#Construyo y ejecuto la tubería
full_pipeline = build_full_pipeline(num_attribs, cat_attribs)
cars_prepared_train = full_pipeline.fit_transform(cars_train)

print(f"Dataset listo para entrenar: {cars_prepared_train.shape}")
```

## 6 Seleccionar y entrenar un modelo

### Regresión lineal

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(cars_prepared_train, cars_labels_train)
```

#### Pruebas

Se seleccionan algunas instancias del conjunto de entrenamiento:

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# 1. Predecimos sobre el mismo set de entrenamiento (para ver si aprendió algo)
cars_predictions = lin_reg.predict(cars_prepared_train)

# 2. Calculamos el error cuadrático medio y su raíz
lin_mse = mean_squared_error(cars_labels_train, cars_predictions)
lin_rmse = np.sqrt(lin_mse)

print(f"Error promedio (RMSE): {lin_rmse}")
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(cars_prepared_train, cars_labels_train)

# Evaluamos de nuevo
cars_predictions = tree_reg.predict(cars_prepared_train)
tree_mse = mean_squared_error(cars_labels_train, cars_predictions)
tree_rmse = np.sqrt(tree_mse)

print(f"Nuevo RMSE (Árbol): {tree_rmse}")
```

##### Cross validation

```python
from sklearn.model_selection import cross_val_score

#Nota: Scikit-Learn espera una función de utilidad (mayor es mejor), 
#por eso uso el negativo del MSE y luego le cambio el signo.
scores = cross_val_score(tree_reg, cars_prepared_train, cars_labels_train,
                         scoring="neg_mean_squared_error", cv=10)

tree_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

#Entreno el Bosque
forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
forest_reg.fit(cars_prepared_train, cars_labels_train)

#Error en entrenamiento
forest_predictions = forest_reg.predict(cars_prepared_train)
forest_rmse = np.sqrt(mean_squared_error(cars_labels_train, forest_predictions))
print(f"RMSE Entrenamiento (Forest): {forest_rmse}")

#Validación Cruzada
forest_scores = cross_val_score(forest_reg, cars_prepared_train, cars_labels_train,
                                scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```
