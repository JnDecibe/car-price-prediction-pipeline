```python
import os
import tarfile
from six.moves import urllib

#Configuración de rutas y URLs:

#DOWNLOAD_ROOT: Dirección base del repositorio de GitHub donde están los datos
DOWNLOAD_ROOT = "https://github.com/JnDecibe/car-price-prediction-pipeline/tree/main/data"

#Selecciono el archivo específico que quiero usar
SELECTED_FILE = "car details v4.csv"

#La URL completa se construye sumando ambas
DOWNLOAD_URL = DOWNLOAD_ROOT + SELECTED_FILE

#CARS_PATH: Ruta local relativa donde guardao el dataset (datasets/cars)
CARS_PATH = os.path.join("datasets", "cars")

#CARS_URL: URL completa para descargar el archivo comprimido (.tgz)
CARS_URL = DOWNLOAD_ROOT + "datasets/cars/cars.tgz"

def fetch_housing_data(cars_url=CARS_URL, cars_path=CARS_PATH):
    """
    Crea el directorio local, descarga el archivo comprimido desde el 
    repositorio y lo extrae para obtener el CSV original.
    """
    #Crea la carpeta 'datasets/housing' si no existe actualmente
    if not os.path.isdir(cars_path):
        os.makedirs(cars_path)
    
    #Define la ruta local completa para el archivo descargado   
    tgz_path = os.path.join(cars_path, "cars.tgz")
    
    #Descarga el archivo desde la URL y lo guarda en tgz_path
    urllib.request.urlretrieve(cars_url, tgz_path)
    
    #Abre el archivo comprimido (.tgz) y lo extrae en la carpeta de destino
    cars_tgz = tarfile.open(tgz_path)
    cars_tgz.extractall(path=cars_path)
    cars_tgz.close() #Cierra el archivo para liberar recursos de memoria
```
