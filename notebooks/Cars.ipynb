```python
import os
import tarfile
from six.moves import urllib

#Configuración de rutas y URLs:

#DOWNLOAD_ROOT: Dirección base del repositorio de GitHub donde están los datos
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"

#HOUSING_PATH: Ruta local relativa donde guardaremos el dataset (datasets/housing)
HOUSING_PATH = os.path.join("datasets", "housing")

#HOUSING_URL: URL completa para descargar el archivo comprimido (.tgz)
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    """
    Crea el directorio local, descarga el archivo comprimido desde el 
    repositorio y lo extrae para obtener el CSV original.
    """
    #Crea la carpeta 'datasets/housing' si no existe actualmente
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    
    #Define la ruta local completa para el archivo descargado   
    tgz_path = os.path.join(housing_path, "housing.tgz")
    
    #Descarga el archivo desde la URL y lo guarda en tgz_path
    urllib.request.urlretrieve(housing_url, tgz_path)
    
    #Abre el archivo comprimido (.tgz) y lo extrae en la carpeta de destino
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close() #Cierra el archivo para liberar recursos de memoria
```
